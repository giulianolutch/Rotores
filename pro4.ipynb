```python
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
csi = 0.05
eta = 2*csi
w = 1
k = 1
m = 1
c = csi*2*sp.sqrt(m*k)
```

```python
def amplitude1(wmin, wmax, csi):
    w = sp.linspace(wmin, wmax, 100*(wmax-wmin))
    x1 = 1/sp.sqrt((k-w**2)**2+(2*csi*w)**2)
    plt.plot(w, x1)
    plt.xlabel('w$_f$/w$_n$')
    plt.ylabel('Xw$_n$²/F$_o$')

def amplitude2(wmin, wmax, csi):
    w = sp.linspace(wmin, wmax, 100*(wmax-wmin))
    x2 = 1/sp.sqrt((k-w**2)**2+(2*eta*k)**2)
    plt.plot(w, x2)
    plt.xlabel('w$_f$/w$_n$')
    plt.ylabel('Xw$_n$²/F$_o$')001
```

```python
amplitude1(wmin = .1, wmax=3., csi= 0.05)
```

```python
interact(amplitude1, wmin=.1, wmax=3., csi=(0.01,.7,0.01))
```

```python
amplitude2(wmin = .1, wmax=3., csi= 0.05)
```
