```python
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
mo = 2
m = 20
e = 0.2
k = (mo*e/m)
wn = 1*2*sp.pi
wr1 = 1*1*sp.pi
wr2 = 10*1*sp.pi
```

```python
r = sp.linspace(wn/wr1,wn/wr2,1000)
```

```python
def f(wr1, wr2, p):
    x = k/sp.sqrt((1-r**2)**2+(2*p*r)**2)
    plt.plot(r,x)
```

```python
interact(f,wr1=1.,wr2=2., p=0.5)
```

```python
r1 = 1*sp.pi
r2 = 2*sp.pi
```

```python
def desbalanceamento(r1, r2, p):
    r = sp.linspace(r1, r2, 1000)
    x = k*r**2/sp.sqrt((1-r**2)**2+(2*p*r))
    plt.plot(r, x)
    
```

```python
interact (desbalanceamento,r1=1.,r2=2.,p=0.5)
```
